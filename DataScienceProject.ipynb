Recommendations can be provided to focus on particular groups to reduce the likelihood of a stroke.

Hypothesis: Age, hypertension and smoking habits are significant predictors influencing the likelihood of experiencing a stroke among adults'

## Data Processing

The data sourced for the project was from Kaggle's 'Stroke Prediction Dataset'. This dataset was chosen due to the range of variables effecting a patients likelihood of having a stroke. 


This is the link to the dataset - (https://www.kaggle.com/datasets/fedesoriano/stroke-prediction-dataset/data)

Data Fields include - 
* ID - Patients unique identifier
* Gender - Patients gender - Male, Female or Other
* Age - Patients age
* Hypertension - 0 if patient does not have hypertension. 1 if patient has hypertension
* Heart_disease - 0 if patient does not have heart disease. 1 if patient has heart disease.
* Ever_married - Patient ever been married - Yes, No
* Work_type - Patients work status. Children, Govt_job, Never worked, Private, Self employed.
* Residence_type - Patient's home address - Rural, Urban.
* Avg_glucose_level - Patient's average glucose level in blood.
* BMI - Patient's Body Mass Index. 
* Smoking_Status - Patients smoking status. Formerly smoked, Never smoked, Smokes, Unknown.
* Stroke - 0 if patient has not had a stroke. 1 if patient has had a stroke.

```python
# Calculate Stroke Occurence vs Marital Status 
# Map categorical values to numerical values
stroke['PatientHadStroke'] = stroke['PatientHadStroke'].map({'Yes': 1, 'No': 0})

# Get the counts for each combination of MaritalStatus and PatientHadStroke
count_combinations = stroke.groupby(['Married', 'PatientHadStroke']).size().reset_index(name='Count')

# Pivot the table for better display and calculations
pivot_table = count_combinations.pivot(index='Married', columns='PatientHadStroke', values='Count').fillna(0)

# Calculate the percentage of 'Had Stroke' cases within each MaritalStatus group
pivot_table['Percentage_Had_Stroke'] = (pivot_table[1] / (pivot_table[0] + pivot_table[1])) * 100

print(pivot_table)
```
