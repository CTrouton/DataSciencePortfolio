Recommendations can be provided to focus on particular groups to reduce the likelihood of a stroke.

Hypothesis: Age, hypertension and smoking habits are significant predictors influencing the likelihood of experiencing a stroke among adults'

## Data Processing

The data sourced for the project was from Kaggle's 'Stroke Prediction Dataset'. This dataset was chosen due to the range of variables effecting a patients likelihood of having a stroke. 


This is the link to the dataset - (https://www.kaggle.com/datasets/fedesoriano/stroke-prediction-dataset/data)

Data Fields include - 
* ID - Patients unique identifier
* Gender - Patients gender - Male, Female or Other
* Age - Patients age
* Hypertension - 0 if patient does not have hypertension. 1 if patient has hypertension
* Heart_disease - 0 if patient does not have heart disease. 1 if patient has heart disease.
* Ever_married - Patient ever been married - Yes, No
* Work_type - Patients work status. Children, Govt_job, Never worked, Private, Self employed.
* Residence_type - Patient's home address - Rural, Urban.
* Avg_glucose_level - Patient's average glucose level in blood.
* BMI - Patient's Body Mass Index. 
* Smoking_Status - Patients smoking status. Formerly smoked, Never smoked, Smokes, Unknown.
* Stroke - 0 if patient has not had a stroke. 1 if patient has had a stroke.

## Data Quality

Following investigation of fields and values within the table, I continue with a data quality audit. I check numerous aspects of the data ensuring it's fit for the intended use within this project. 
There are key components within data quality checking which are: 
* Accuracy - check for errors/inconsistencies
* Completeness - check for missing values
* Consistency - check for uniformity and coherence within the data
* Timeliness - data is a snapshot in time so unable to check timeliness
* Relevance - data is relevant to project 
* Validity - check data conforms to structure and format expected
* Reliability - data is reliable and can be trusted for making decisions


```python
# Count null values in each column
null_counts = stroke.isnull().sum()
display(null_counts)
```

```python
# Remove missing/null values from data frame.
stroke = stroke.dropna()
```

```python
# Check NULL values after removal #
null_counts = stroke.isnull().sum()
display(null_counts)
```

```python
# Filtering out rows where gender is not 'Other'
stroke = stroke[stroke['gender'] != 'Other']

# Counting occurrences of each gender category
gender_counts = stroke['gender'].value_counts()
print(gender_counts)
```

```python
# Tidy column names 

stroke = stroke.rename(columns={'ever_married': 'Married'})
stroke = stroke.rename(columns={'id':'PatientID'})
stroke = stroke.rename(columns={'gender':'Gender'})
stroke = stroke.rename(columns={'age':'PatientsAge'})
stroke = stroke.rename(columns={'hypertension':'Hypertension'})
stroke = stroke.rename(columns={'heart_disease':'HeartDisease'})
stroke = stroke.rename(columns={'work_type':'WorkType'})
stroke = stroke.rename(columns={'Residence_type':'ResidenceType'})
stroke = stroke.rename(columns={'avg_glucose_level':'AverageGlucoseLevel'})
stroke = stroke.rename(columns={'bmi':'BodyMassIndex'})
stroke = stroke.rename(columns={'smoking_status':'SmokingStatus'})
stroke = stroke.rename(columns={'stroke':'PatientHadStroke'})
stroke
```
